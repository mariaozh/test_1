**Задание:**


**1.   Создайте класс Chocolate (для определения шоколадного батончика), принимающий следующие аргументы при инициализации: цена батончика (тип - float), добавка к шоколаду (необязательный аргумент, тип - str) - 2 балла**

**2.   При инициализации необходимо проверять, чтобы цена батончика была больше нуля, если цена меньше нуля необходимо вызывать исключение ValueError - 2 балла**

**3.   Реализуйте метод show_my_chocolate, выводящий на печать "Шоколадный батончик, добавка: {ДОБАВКА}" в случае наличия добавки, а иначе должна отобразиться следующая фраза: "Обычный шоколадный батончик" - 2 балла**

**4.   Переопределите метод __str__, который при применении функции print() к объекты класса Chocolate должен возвращать строку "Шоколадный батончик, <ЦЕНА> рублей" - 2 балла**

**5. Продемонстрируйте использование соответствующего функционала вашего класса, создав несколько объектов Chocolate с различными аргументами - 2 балла**

```python
class Chocolate:
    def __init__(self, price, add='No add'):  # определяем конструктор класса
      if  price > 0: #проверяем, чтобы цена батончика была больше нуля
        self.price = float(price)  # присваиваем атрибуту price значение параметра price
      else:
        raise ValueError('Меньше или равно 0!') #Вызов исключения
      self.add = str(add) # присваиваем атрибуту age значение параметра age, атрибут не обязательный, значение по умолчанию "No add"

    def display_info(self):
      print(f"Цена: {self.price}, добавка к шоколадке: {self.add}")

    def show_my_chocolate(self):
        if self.add != 'No add':
          print(f"Шоколадный батончик, добавка: {self.add}")
        else:
          print(f"Обычный шоколадный батончик")

    def __str__ (self):
      return f"Шоколадный батончик, {self.price}  рублей"

# Ввод данных о батончиках
choco_1 = Chocolate(100,'орехи')  # создаем объект класса Chocolate
choco_2 = Chocolate(20,'маршмэллоу')  # создаем объект класса Chocolate
choco_3 = Chocolate(286.3,'печенье')  # создаем объект класса Chocolate
choco_4 = Chocolate(89,'кокос')  # создаем объект класса Chocolate
choco_5 = Chocolate(89) # создаем объект класса Chocolate, без добавки


# Проверяем
print ('\033[32mТЕСТ ЛИЧНЫЙ\033[0m')
print ('')
print('\033[32mВыводим параметры отдельно:\033[0m')
print (choco_1.price)
print (choco_1.add)
print (choco_2.price)
print (choco_2.add)
print (choco_5.price)
print (choco_5.add)
print ('')

print ('\033[32mВыводим методом display_info:\033[0m')
choco_1.display_info()
choco_2.display_info()
choco_3.display_info()
choco_4.display_info()
choco_5.display_info()
print ('')
print ('')

print ('\033[36mРЕЗУЛЬТАТЫ ПО ЗАДАНИЮ 1/2\033[0m')
print ('')
print ('\033[32mВыводим методом show_my_chocolate:\033[0m')
choco_5.show_my_chocolate()
choco_1.show_my_chocolate()
choco_2.show_my_chocolate()
print ('')

print ('\033[32mВыводим методом __str__:\033[0m')
print (choco_1.__str__())
print (choco_5.__str__())
print (choco_2.__str__())
print ('')
```
